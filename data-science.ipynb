# Retrieve a dataset from Web/blogs/Social Media using Python using BeautifulSoup

````python
import requests
from bs4 import BeautifulSoup

# URL of the webpage to scrape
url = 'https://www.hackerearth.com/practice/notes/python-diaries-chapter-1/'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content of the webpage using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all the article titles using the appropriate HTML tags and classes
    article_titles = soup.find_all('h1')
    
    # Extract and print the titles
    print("Article Titles:",end="\n\n")
    for title in article_titles:
        print(title.text)
    
    # Find all the code snippets using the appropriate HTML tags and classes
    code_snippets = soup.find_all('code')
    
    # Extract and print the code snippets
    print("\nCode Snippets:")
    for code in code_snippets:
        print(code.text)
else:
    print('Failed to retrieve the webpage.')
````

# Implement Euclidean distance using Numpy in Python.

![](img/euclids-ans.webp)


````python
""" 
Euclidean distance is a measure of the straight-line distance between two points in a Euclidean space. 
In a two-dimensional space, the Euclidean distance between two points (x1, y1) and (x2, y2) is calculated using the formula:

distance = sqrt((x2 - x1)^2 + (y2 - y1)^2) 
"""

import numpy as np


def euclidean_distance(point1, point2):
    # Convert the points to NumPy arrays for vectorized operations
    point1 = np.array(point1)
    point2 = np.array(point2)

    # Calculate the squared differences between corresponding coordinates
    squared_diffs = (point2 - point1)**2

    # Sum the squared differences and take the square root to get the distance
    distance = np.sqrt(np.sum(squared_diffs))

    return distance


# Example usage
point_a = (-2, 1)
point_b = (2, -1)

distance = euclidean_distance(point_a, point_b)
print("Euclidean distance between point A and point B:", distance)
````

## Determine the probability of a hypothesis with prior knowledge using Bayes' theorem using Numpy in Python

![](img/bayes.png)


````python
"""
 Baye's therorem is used in probability and statistics to update the probability of a hypothesis based on new evidence
 Formula: P(A|B) = (P(B|A) * P(A)) / P(B)
 P(A|B) is the probability of hypothesis A given evidence B (posterior probability).
 P(B|A) is the probability of evidence B given hypothesis A (likelihood).
 P(A) is the prior probability of hypothesis A.
 P(B) is the probability of evidence B.
"""

import numpy as np


def bayes_theorem(prior_prob, likelihood, evidence_prob):
    # Calculate the posterior probability using Bayes' theorem formula
    posterior_prob = (likelihood * prior_prob) / evidence_prob
    return posterior_prob


# Example usage
prior_probability = 0.3       # Prior probability of the hypothesis
likelihood = 0.7             # Likelihood of the evidence given the hypothesis
evidence_probability = 0.5   # Probability of the evidence

posterior_probability = bayes_theorem(
    prior_probability, likelihood, evidence_probability)
print("Posterior probability:", posterior_probability)
````

## The Manhattan distance as the sum of absolute differencesManhattanDistance [{a, b, c}, {x, y, z}] Abs [a − x] + Abs [b − y] + Abs [c − z]. Find the Manhattan distance using Numpy in Python.

![](./img/manhat-dist-ans.webp)


````python
# The Manhattan distance is the sum of the absolute differences between the corresponding coordinates of two points.

import numpy as np

def manhattan_distance(point1, point2):
    # Convert the points to NumPy arrays for vectorized operations
    point1 = np.array(point1)
    point2 = np.array(point2)

    # Calculate the absolute differences between corresponding coordinates
    absolute_diffs = np.abs(point2 - point1)

    # Sum the absolute differences to get the Manhattan distance
    distance = np.sum(absolute_diffs)

    return distance


# Example usage
point_a = (1, 1, 0)
point_b = (4, 3, 0)

distance = manhattan_distance(point_a, point_b)
print("Manhattan distance between point A and point B:", distance)
````

## The Jaccard similarity is calculated by dividing the number of observations in both sets by the number of observations in either set.the Jaccard similarity can be computed as the size of the intersection divided by the size of the union of two sets. This can be written in set notation using intersectionand unions of two sets:J(X,Y) = |X∩Y| / |X∪Y|

```
   X={ 0,1,2,5,6}
   Y={0,2,3,4,5,7,9} Implement with numpy in python.
```

````python
"""
we first convert the sets `X` and `Y` into lists using `list()`, 
then use `np.intersect1d()` to find the intersection of the two sets and 
`np.union1d()` to find the union. 
Finally, we calculate the Jaccard similarity by dividing the size of the intersection by the size of the union.
"""

import numpy as np

X = {0, 1, 2, 5, 6}
Y = {0, 2, 3, 4, 5, 7, 9}

intersection_size = len(np.intersect1d(list(X), list(Y)))
union_size = len(np.union1d(list(X), list(Y)))

jaccard_similarity = intersection_size / union_size
print("Jaccard Similarity:", jaccard_similarity)
````

## Below Figure shows two vectors x and y and the angle θ between them. Determine the geometric formula of dot product using numpy in Python which is defined as:

![](./img/6th.png)
![](./img/7th.jpg)


geometric definition of the dot product between two vectors. The dot product of two vectors x and y can be calculated using the formula:

```py
x ⋅ y = |x| * |y| * cos(θ)
```

> Where:

> + x ⋅ y is the dot product of vectors x and y.  
> + |x| is the magnitude (length) of vector x.  
> + |y| is the magnitude (length) of vector y.  
> + θ is the angle between vectors x and y.  

````python
import numpy as np

x1 = int(input("Enter x1:"))
x2 = int(input("Enter x2:"))
x3 = int(input("Enter x3:"))
y1 = int(input("Enter y1:"))
y2 = int(input("Enter y2:"))
y3 = int(input("Enter y3:"))
    
x = np.array([x1, x2, x3])  # Replace x1, x2, x3 with the components of vector x
y = np.array([y1, y2, y3])  # Replace y1, y2, y3 with the components of vector y

dot_product = np.dot(x, y)
magnitude_x = np.linalg.norm(x)
magnitude_y = np.linalg.norm(y)

cos_theta = dot_product / (magnitude_x * magnitude_y)
theta = np.arccos(cos_theta)

print("Dot Product:", dot_product)
print("Magnitude of x:", magnitude_x)
print("Magnitude of y:", magnitude_y)
print("Angle theta in radians:", theta)
````
